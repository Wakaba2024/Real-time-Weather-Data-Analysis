```python
import requests
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# API keys
weather_api_key = 
open_weather_api_key = 

# Cities to collect data for
cities = ["Nairobi", "Mombasa", "Kisumu", "Eldoret"]

# Data collection function
def collect_weather_data(city):
    # WeatherAPI request
    weather_url = f"http://api.weatherapi.com/v1/current.json?key={weather_api_key}&q={city}"
    weather_response = requests.get(weather_url).json()
    
    # Extract data from WeatherAPI
    weather_api_data = {
        "city": city,
        "last_updated_epoch": weather_response['current']['last_updated_epoch'],
        "last_updated": weather_response['current']['last_updated'],
        "temperature_c": weather_response['current']['temp_c'],
        "humidity": weather_response['current']['humidity'],
        "wind_speed_kph": weather_response['current']['wind_kph'],
        "wind_degree": weather_response['current']['wind_degree'],
        "pressure_in": weather_response['current']['pressure_in'],
        "precip_in": weather_response['current']['precip_in'],
        "cloud": weather_response['current']['cloud'],
        "feelslike_c": weather_response['current']['feelslike_c'],
        "gust_kph": weather_response['current']['gust_kph'],
        "windchill_c": weather_response.get('current', {}).get('windchill_c'),
        "heatindex_c": weather_response.get('current', {}).get('heatindex_c'),
        "dewpoint_c": weather_response.get('current', {}).get('dewpoint_c'),
        "timestamp": datetime.now()
    }

    # OpenWeatherMap request
    open_weather_url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={open_weather_api_key}&units=metric"
    open_weather_response = requests.get(open_weather_url).json()
    
    # Extract data from OpenWeatherMap
    open_weather_data = {
        "temperature_c_open": open_weather_response['main']['temp'],
        "humidity_open": open_weather_response['main']['humidity'],
        "wind_speed_mps": open_weather_response['wind']['speed']
    }

    # Combine data from both sources
    weather_data = {**weather_api_data, **open_weather_data}
    return weather_data

# Collect data for all cities and store in a DataFrame
weather_data_list = [collect_weather_data(city) for city in cities]
df_weather = pd.DataFrame(weather_data_list)
```

# STORE THE DATA INTO A MYSQL DATABASE

```python
# Function to insert weather data into the MySQL database
def store_data_in_db(data, engine):
    # Create a DataFrame with the data
    df = pd.DataFrame([data])
    
    # Store the data in the MySQL table 'weather'
    df.to_sql('weather', con=engine, if_exists='append', index=False)
    print(f"Weather data for {data['city']} inserted successfully.")

# Fetch and store data for each city
for city in cities:
    try:
        weather_data = collect_weather_data(city)
        store_data_in_db(weather_data, engine)
    except Exception as e:
        print(f"Error fetching or storing data for {city}: {e}")
```

```python
# Define the MySQL connection string
db_username = ''
db_password = ''
db_host = 'localhost'
db_name = 'weather_data'

# Create a connection engine using SQLAlchemy
connection_string = f'mysql+pymysql://{db_username}:{db_password}@{db_host}/{db_name}'
engine = create_engine(connection_string)

# Load data into a DataFrame using the engine
query = "SELECT * FROM weather"
df_weather = pd.read_sql(query, engine)
```

```python
df_weather
```

# CLEAN THE DATA

```python
# Convert 'last_updated' and 'timestamp' columns to datetime
df_weather['last_updated'] = pd.to_datetime(df_weather['last_updated'])
df_weather['timestamp'] = pd.to_datetime(df_weather['timestamp'])

# Fill missing values using forward fill method
df_weather.ffill(inplace=True)

# Drop duplicates based on city and timestamp
df_weather.drop_duplicates(subset=['city', 'timestamp'], inplace=True)

```

#  Exploratory Data Analysis(EDA)

```python
# I will perform EDA on the cleaned data. This includes analyzing temperature trends, rainfall patterns, and weather variables
```

# Temperature Trends Visualization

```python
# Line plot of temperature trends for each city.
```

```python
# Set plot style
sns.set(style="whitegrid")

# Plot temperature trends for each city
plt.figure(figsize=(10, 6))
for city in df_weather['city'].unique():
    city_data = df_weather[df_weather['city'] == city]
    plt.plot(city_data['timestamp'], city_data['temperature_c'], label=city)

plt.title('Temperature Trends in Kenyan Cities')
plt.xlabel('last_updated')
plt.ylabel('city')
plt.legend(title='City')
plt.show()
```

# Rainfall Patterns Visualization

```python
# Query to fetch rainfall data
query = "SELECT city, timestamp, precip_in FROM weather"
df_rainfall = pd.read_sql(query, engine)

# Convert 'timestamp' column to datetime
df_rainfall['timestamp'] = pd.to_datetime(df_rainfall['timestamp'])

# Extract date from timestamp
df_rainfall['date'] = df_rainfall['timestamp'].dt.date

# Group by city and date, summing the precipitation
df_rainfall_grouped = df_rainfall.groupby(['city', 'date'], as_index=False).agg({'precip_in': 'sum'})

# Set plot style
sns.set(style="whitegrid")

# Create a boxplot for rainfall patterns by city
plt.figure(figsize=(12, 6))
sns.boxplot(x='city', y='precip_in', data=df_rainfall_grouped)

# Formatting the plot
plt.title('Rainfall Patterns in Kenyan Cities')
plt.xlabel('City')
plt.ylabel('Total Precipitation (inches)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

# Weather Variables Visualization

# HUMIDITY TRENDS

```python
# Using a line plot to show humidity trends over time for each city.
```

```python
# Query to fetch humidity data
query = "SELECT city, timestamp, humidity FROM weather"
df_humidity = pd.read_sql(query, engine)

# Convert 'timestamp' column to datetime
df_humidity['timestamp'] = pd.to_datetime(df_humidity['timestamp'])

# Set plot style
sns.set(style="whitegrid")

# Plot humidity trends
plt.figure(figsize=(12, 6))
for city in df_humidity['city'].unique():
    city_data = df_humidity[df_humidity['city'] == city]
    plt.plot(city_data['timestamp'], city_data['humidity'], label=city)

plt.title('Humidity Trends in Kenyan Cities')
plt.xlabel('Timestamp')
plt.ylabel('Humidity (%)')
plt.legend(title='City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# HEAT PATTERNS

```python
#  Using a scatter plot to visualize the relationship between the heatindex_c vs temperature and humidity.
```

```python
# Query to fetch temperature and humidity data
query = "SELECT city, temperature_c, heatindex_c, humidity FROM weather"
df_heat = pd.read_sql(query, engine)

# Set plot style
sns.set(style="whitegrid")

# Create a scatter plot for temperature vs humidity
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_heat, x='temperature_c', y='humidity', hue='city', style='city', s=100)

plt.title('Temperature vs Humidity in Kenyan Cities')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.legend(title='City')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,2)
sns.scatterplot(data = df_heat, x='temperature_c', y='heatindex_c', ax = ax[0])
sns.scatterplot(data = df_heat, x='humidity', y='heatindex_c', ax = ax[1])       
```

# WIND PATTERNS

```python
# Using a bar plot to compare average wind speeds across cities.
```

```python
# Query to fetch wind speed data
query = "SELECT city, wind_speed_kph FROM weather"
df_wind = pd.read_sql(query, engine)

# Calculate average wind speed per city
avg_wind_speed = df_wind.groupby('city', as_index=False).agg({'wind_speed_kph': 'mean'})

# Set plot style
sns.set(style="whitegrid")

# Create a bar plot for average wind speed
plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='wind_speed_kph', data=avg_wind_speed)

plt.title('Average Wind Speed in Kenyan Cities')
plt.xlabel('City')
plt.ylabel('Average Wind Speed (kph)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# CLOUD COVER PATTERNS

```python
#  Using a box plot to display the distribution of cloud cover for each city.
```

```python
# Query to fetch cloud data
query = "SELECT city, cloud FROM weather"
df_cloud = pd.read_sql(query, engine)

# Set plot style
sns.set(style="whitegrid")

# Create a box plot for cloud cover by city
plt.figure(figsize=(12, 6))
sns.boxplot(x='city', y='cloud', data=df_cloud)

# Formatting the plot
plt.title('Cloud Cover Patterns in Kenyan Cities')
plt.xlabel('City')
plt.ylabel('Cloud Cover (%)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
